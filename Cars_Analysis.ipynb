## Key Findings

- **Asian manufacturers dominate by volume** — the dataset contains significantly more Asian car models than European or American, reflecting the global shift toward Japanese, South Korean, and Chinese automakers.

- **European cars command the highest average prices** — brands like BMW, Mercedes-Benz, and Jaguar push the average MSRP for European vehicles up to premium levels, while American and Asian brands show wider price distributions.

- **Engine size and cylinders are strong price predictors** — a direct positive correlation exists between cylinder count and horsepower; cars with 8 cylinders average nearly 2× the horsepower of 4-cylinder vehicles.

- **Horsepower scales dramatically with engine size** — 8-cylinder cars (5.0+ liters) average 400+ HP, while 4-cylinder cars (2.0-2.5 liters) average 160-180 HP, creating distinct performance tiers.

- **Sedan models are the most fuel-efficient** — averaging 22 MPG city / 29 MPG highway, while SUVs average only 17 city / 23 highway due to their weight and aerodynamic characteristics.

- **SUVs are significantly heavier** — averaging 4,600+ lbs compared to sedans at 3,200 lbs, explaining both their lower fuel efficiency and their price premium for a given engine size.

- **Higher horsepower commands premium pricing** — a clear positive correlation shows that each additional 50 HP correlates with a £5,000-£8,000 price increase in MSRP across comparable vehicle classes.

- **Fuel efficiency and performance are inversely related** — the highest HP vehicles consistently achieve the lowest MPG ratings, forcing buyers to choose between power and economy.

- **All-wheel drive vs Front-wheel drive impacts both price and efficiency** — all-wheel drive vehicles are typically £2,000-£4,000 more expensive and 2-3 MPG less efficient due to drivetrain complexity.
